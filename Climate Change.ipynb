Emissioni Globali di CO2

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file CSV
file_path = r'C:\Users\Luca_\Python Epicode\Progetto Climate Change\Sistemati\CO2\emissioni annuali di co2 per paese.csv'
df = pd.read_csv(file_path)

# Filtra i dati per l'entità "World"
world_emissions = df[df['Entity'] == 'World']

# Crea un grafico a linee delle emissioni totali di CO₂
plt.figure(figsize=(12, 6))
plt.plot(world_emissions['Year'], world_emissions['Annual CO₂ emissions'], label='Emissioni totali CO₂', color='green')
plt.title('Andamento delle Emissioni Totali di CO₂ nel Mondo dal 1750 al 2022')
plt.xlabel('Anno')
plt.ylabel('Emissioni di CO₂ (tonnellate)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import plotly.express as px

# Filtrare i dati per il mondo
world_data = df[df['Entity'] == 'World']

# Creare un grafico interattivo con Plotly
fig = px.line(world_data, x='Year', y='Annual CO₂ emissions', 
              title='Emissioni Globali di CO2 dal 1750 al 2022',
              labels={'Annual CO₂ emissions': 'Emissioni di CO2 (tonnellate)', 'Year': 'Anno'})

# Mostrare il grafico
fig.show()
```
