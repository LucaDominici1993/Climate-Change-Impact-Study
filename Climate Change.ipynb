Emissioni Globali di CO2

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file CSV
file_path = r'C:\Users\Luca_\Python Epicode\Progetto Climate Change\Sistemati\CO2\emissioni annuali di co2 per paese.csv'
df = pd.read_csv(file_path)

# Filtra i dati per l'entità "World"
world_emissions = df[df['Entity'] == 'World']

# Crea un grafico a linee delle emissioni totali di CO₂
plt.figure(figsize=(12, 6))
plt.plot(world_emissions['Year'], world_emissions['Annual CO₂ emissions'], label='Emissioni totali CO₂', color='green')
plt.title('Andamento delle Emissioni Totali di CO₂ nel Mondo dal 1750 al 2022')
plt.xlabel('Anno')
plt.ylabel('Emissioni di CO₂ (tonnellate)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import plotly.express as px

# Filtrare i dati per il mondo
world_data = df[df['Entity'] == 'World']

# Creare un grafico interattivo con Plotly
fig = px.line(world_data, x='Year', y='Annual CO₂ emissions', 
              title='Emissioni Globali di CO2 dal 1750 al 2022',
              labels={'Annual CO₂ emissions': 'Emissioni di CO2 (tonnellate)', 'Year': 'Anno'})

# Mostrare il grafico
fig.show()
```

anomalie temperature

```python
import pandas as pd
import plotly.express as px

# Carica il file CSV in un DataFrame
data_path = r'C:\Users\Luca_\Python Epicode\Progetto Climate Change\Sistemati\Temperature\Anomalie mensili tempertura dal 1880 al 2022.csv'
monthly_anomalies = pd.read_csv(data_path)

# Assicurati che la colonna 'Date' sia in formato datetime
monthly_anomalies['Date'] = pd.to_datetime(monthly_anomalies['Date'])

# Crea il grafico a linee con Plotly
fig = px.line(monthly_anomalies, x='Date', y='Mean', color='Source',
              title='Monthly Temperature Anomalies from GISTEMP and GCAG',
              labels={'Mean': 'Temperature Anomaly (°C)', 'Date': 'Date'},
              template='plotly_white')

# Personalizza il grafico
fig.update_traces(mode='lines+markers', hoverinfo='x+y+text', marker=dict(size=4))
fig.update_layout(legend_title_text='Data Source')
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightPink')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightBlue')

# Mostra il grafico
fig.show()

# Salva il grafico come file HTML per la visualizzazione esterna
fig.write_html('monthly_temperature_anomalies_interactive.html')
```

```python
import pandas as pd
import plotly.express as px

# Carica il file CSV in un DataFrame
annual_data_path = r'C:\Users\Luca_\Python Epicode\Progetto Climate Change\Sistemati\Temperature\Anomalie annuali temperatura dal 1880 al 2022.csv'  # Sostituisci con il percorso al tuo file
annual_anomalies = pd.read_csv(annual_data_path)

# Assicurati che la colonna 'Year' sia in formato datetime
annual_anomalies['Year'] = pd.to_datetime(annual_anomalies['Year'], format='%Y')

# Crea il grafico a linee con Plotly
fig_annual = px.line(annual_anomalies, x='Year', y='Mean', color='Source',
                     title='Annual Temperature Anomalies from GISTEMP and GCAG',
                     labels={'Mean': 'Temperature Anomaly (°C)', 'Year': 'Year'},
                     template='plotly_white')

# Personalizza il grafico
fig_annual.update_traces(mode='lines+markers', hoverinfo='x+y+text', marker=dict(size=4))
fig_annual.update_layout(legend_title_text='Data Source')
fig_annual.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightPink')
fig_annual.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightBlue')

# Mostra il grafico
fig_annual.show()

# Salva il grafico come file HTML per la visualizzazione esterna
fig_annual.write_html('annual_temperature_anomalies_interactive.html')
```
